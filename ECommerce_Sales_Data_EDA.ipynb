```python
all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']
all_data.shape
```

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
all_data.head()
```

```python
# Slicing first two characters from Order Date for month and creating new column
all_data['Month'] = all_data['Order Date'].str[0:2]
```

```python
#Convert Month from str to int32
all_data['Month'] = all_data['Month'].astype('int32')
all_data.head()
```

```python
#creating a total sales column
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

```python
#adding a city column
# Let's use the .apply()
def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return (address.split(',')[2]).split(' ')[1]

all_data['City'] = all_data['Purchase Address'].apply(lambda x: get_city(x)+' ('+get_state(x)+')')

all_data.head()
```

```python
results = all_data.groupby('Month').sum() 
results
# It only sums up the columns with numeric datatype!!
```

```python
import  matplotlib.pyplot as plt
```

```python
#plotting a chart to determine total sales monthwise
months = range(1,13)
plt.figure(figsize=(9,4))
plt.bar(months,results['Sales'])
plt.xticks(months)
plt.ylabel('Total Sales in a month in USD ($)')
plt.xlabel('Month Number')
plt.grid(axis='y')
plt.show()
```

```python
#Citywise Sales
results = all_data.groupby('City').sum() 
results
```

```python
cities = results.index
plt.figure(figsize=(9,4))
plt.bar(cities,results['Sales'])
plt.xticks(cities, rotation='vertical', size=8)
plt.ylabel('Total Sales in a city in USD ($)')
plt.xlabel('City Name')
plt.grid(axis='y')
plt.show()
```

```python
#converting Order Date to datetime
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
```

```python
all_data.head()
```

```python
#creating hour and minute columns
all_data['Hour'] = all_data['Order Date'].dt.hour
all_data['Minute'] = all_data['Order Date'].dt.minute
all_data.head()
```

```python
#grouping the data based on hour
results = all_data.groupby(['Hour']).count()
results
```

```python
#plotting sales data against hour of day to determine best times to display ads based on sales activity
hours = results.index
plt.plot(hours, results['Sales'])
plt.xticks(hours)
plt.xlabel('Hour of the Day')
plt.ylabel('Total Number of Sales in an Hour')
plt.grid()
plt.show()
```

From the chart above, it is clearly evident that the peaks in shopping occur around 12pm (1200 hrs) and 7pm (1900 hrs) across the entire 10 US cities.

Here, by carefully observing the data, we can say that if the Order ID of two or more rows match, the corresponding Products were sold together.

```python
# Keeping only the ones which have duplicated Order ID
df = all_data[all_data['Order ID'].duplicated(keep=False)]

# Joining all the Products with same Order ID by ',' and storing in 'Grouped' column
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))

# Take only the Order ID and Grouped columns and drop duplicates
df = df[['Order ID', 'Grouped']].drop_duplicates()
df.head(100)
```

```python
#counting the pairs of products that are sold together
from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

for key, value in count.most_common(10):
    print(key, value)
```
