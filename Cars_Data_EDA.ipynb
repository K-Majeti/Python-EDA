```python
# Selecting only numeric columns before computing quantiles
numeric_cols = df.select_dtypes(include=['int64', 'float64'])
Quant1 = numeric_cols.quantile(0.25)
Quant3 = numeric_cols.quantile(0.75)
IQR = Quant3 - Quant1

# Output the quantiles and IQR
print("First Quantile:\n", Quant1)
print("Third Quantile:\n", Quant3)
print("Interquartile Range:\n", IQR)
```

```python
# Create a copy of the dataframe
df_cleaned = df.copy()

# Initialize total_outliers
total_outliers = 0

# Loop over each column
for col in numeric_cols.columns:
    Q1 = Quant1[col]
    Q3 = Quant3[col]
    IQR = Q3 - Q1

    # Define the acceptable range
    lower_bound = Q1 - 1.5*IQR
    upper_bound = Q3 + 1.5*IQR

    # Identify outliers
    outliers = df_cleaned[(df_cleaned[col] < lower_bound) | (df_cleaned[col] > upper_bound)]

    # Print the number of outliers in each column
    print(f"Number of outliers in {col}: {len(outliers)}")

    # Add the number of outliers to total_outliers
    total_outliers += len(outliers)

    # Remove outliers
    df_cleaned = df_cleaned[(df_cleaned[col] >= lower_bound) & (df_cleaned[col] <= upper_bound)]

# Print the total number of outliers
print(f"Total number of outliers: {total_outliers}")
```
