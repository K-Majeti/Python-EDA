```python
# Selecting only numeric columns before computing quantiles
numeric_cols = df.select_dtypes(include=['int64', 'float64'])
Quant1 = numeric_cols.quantile(0.25)
Quant3 = numeric_cols.quantile(0.75)
IQR = Quant3 - Quant1

# Output the quantiles and IQR
print("First Quantile:\n", Quant1)
print("Third Quantile:\n", Quant3)
print("Interquartile Range:\n", IQR)
```

```python
# Create a copy of the dataframe
df_cleaned = df.copy()

# Initialize total_outliers
total_outliers = 0

# Loop over each column
for col in numeric_cols.columns:
    Q1 = Quant1[col]
    Q3 = Quant3[col]
    IQR = Q3 - Q1

    # Define the acceptable range
    lower_bound = Q1 - 1.5*IQR
    upper_bound = Q3 + 1.5*IQR

    # Identify outliers
    outliers = df_cleaned[(df_cleaned[col] < lower_bound) | (df_cleaned[col] > upper_bound)]

    # Print the number of outliers in each column
    print(f"Number of outliers in {col}: {len(outliers)}")

    # Add the number of outliers to total_outliers
    total_outliers += len(outliers)

    # Remove outliers
    df_cleaned = df_cleaned[(df_cleaned[col] >= lower_bound) & (df_cleaned[col] <= upper_bound)]

# Print the total number of outliers
print(f"Total number of outliers: {total_outliers}")
```

### Data Visualization

```python
# Plotting a Histogram
df_cleaned.Make.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Number of cars by make")
plt.ylabel('Number of cars')
plt.xlabel('Make');
```

```python
#Using label encoding to convert categorical data to numerical data
from sklearn.preprocessing import LabelEncoder

# Create a copy of the dataframe
df_encoded = df_cleaned.copy()

# Apply LabelEncoder to each column
le = LabelEncoder()
for col in df_encoded.columns:
    df_encoded[col] = le.fit_transform(df_encoded[col])

# Calculate the correlation matrix
c = df_encoded.corr()

# Plot the correlation matrix
plt.figure(figsize=(10,5))
sns.heatmap(c, cmap="BrBG", annot=True)
plt.show()
```

```python
# Create a new column 'Avg_MPG' which is the average of 'MPG-H' and 'MPG-C'
df_cleaned['Avg_MPG'] = (df_cleaned['MPG-H'] + df_cleaned['MPG-C']) / 2

# Plotting the relation of HP and Avg_MPG
plt.figure(figsize=(10,5))
plt.scatter(df_cleaned['HP'], df_cleaned['Avg_MPG'], label='Avg_MPG')

plt.xlabel('HP')
plt.ylabel('Avg_MPG')
plt.title('HP vs Avg_MPG')
plt.legend()
plt.show()
```

```python
df_cleaned['Avg_MPG'] = (df_cleaned['MPG-H'] + df_cleaned['MPG-C']) / 2
# Group the data by 'Make' and calculate the mean of 'Avg_MPG' for each 'Make'
df_grouped = df_cleaned.groupby('Make')['Avg_MPG'].mean()

# Plotting the comparison of average MPG by 'Make'
plt.figure(figsize=(10,5))
df_grouped.sort_values().plot(kind='barh')

plt.xlabel('Average MPG')
plt.ylabel('Make')
plt.title('Comparison of Average MPG by Make')
plt.show()
```

```python
# Plotting the relation of Cylinders and Avg_MPG
df_cleaned['Avg_MPG'] = (df_cleaned['MPG-H'] + df_cleaned['MPG-C']) / 2

plt.figure(figsize=(10,5))
sns.boxplot(x=df_cleaned['Cylinders'], y=df_cleaned['Avg_MPG'])

plt.xlabel('Cylinders')
plt.ylabel('Avg_MPG')
plt.title('Cylinders vs Avg_MPG')
plt.show()
```

```python
# Group the data by 'Make' and calculate the mean of 'Price' for each 'Make'
df_grouped = df_cleaned.groupby('Make')['Price'].mean()

# Plotting the comparison of average MPG by 'Make'
plt.figure(figsize=(10,5))
df_grouped.sort_values().plot(kind='barh')

plt.xlabel('Average Price')
plt.ylabel('Make')
plt.title('Comparison of Average Price by Make')
plt.show()
```

```python
# Group the data by 'Year' and calculate the mean of 'Price' for each year
df_grouped = df_cleaned.groupby('Year')['Price'].mean()

# Plotting the evolution of average prices (MSRP) over the years
plt.figure(figsize=(10,5))
plt.plot(df_grouped.index, df_grouped, label='Average Price')

plt.xlabel('Year')
plt.ylabel('Average Price')
plt.title('Evolution of Average Price over the Years')
plt.legend()
plt.show()
```

# Summary of Findings and Recommendations

## Car Market Analysis Summary

### Number of Cars by Make
- 📊 **Chevrolet, Ford, and Volkswagen** are the most common car makes, each with a significant presence in the dataset.
- 🚗 There is a wide variety of car makes, but only a few dominate the market according to the dataset.
- 🌟 **Insight**: Manufacturers with fewer cars could be niche players or luxury brands, which could be explored further.

### Correlation Heatmap Insights
- 🤔 **Cylinders and horsepower (HP)** have a strong positive correlation, indicating that more cylinders usually mean more power.
- 💡 The **price** has a noticeable positive correlation with HP and cylinders, suggesting that higher performance may command a higher price.
- 🔄 **Drive mode** shows some correlation with MPG-H (Highway Miles per Gallon), potentially for efficiency in specific driving conditions.

### Horsepower vs. Average MPG
- 📉 There's a trend indicating that as **horsepower increases**, the **average miles per gallon (MPG) tends to decrease**.
- ⛽ This trade-off highlights the balance between performance and fuel efficiency in vehicles.

### Comparison of Average MPG by Make
- 🏆 **FIAT, Alfa Romeo, and Scion** lead with the highest average MPG, which could appeal to cost and environmentally conscious consumers.
- 🚀 The data might suggest a focus on fuel efficiency among these makes.

### Cylinders vs. Average MPG
- 🎚️ Vehicles with **4 cylinders** tend to have higher average MPG, while those with **8 cylinders** have lower MPG, reaffirming the performance-efficiency trade-off.
- 🔄 The wide spread in MPG for cars with **6 cylinders** could indicate a range of performance options within that category.

### Comparison of Average Price by Make
- 💲 **Maserati, Porsche, and Lotus** appear to be the most expensive makes on average, which may indicate their luxury status.
- 💵 There is a wide range of average prices, indicating a diverse car market catering to different consumer segments.

### Evolution of Average Price over the Years
- 📈 The average price of cars shows an upward trend over the years, suggesting either inflation, a shift towards more expensive vehicles, or improvements in car quality and features.
- 🧐 A significant jump in price around the early 2000s could warrant further investigation to understand market dynamics at that time.

### Evolution of Average MPG over the Years
- 🌱 There's an overall increasing trend in MPG over the years, indicating improvements in fuel efficiency or a shift towards more fuel-efficient vehicles.
- 📊 Some fluctuations in MPG are noticeable; identifying the causes could provide additional insights into market changes or technological advancements.
